```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data/Seasons_stats_complete.csv')
```

```python
df
```

```python
df = df[df['Year']>2015]
```

```python
df = df.sort_values(['Year', 'Player'])
```

```python
stats_2016 = df[df['Year']==2016]
stats_2016 = stats_2016.drop_duplicates('Player')
stats_2016
```

```python
stats_2017 = df[df['Year']==2017]
stats_2017 = stats_2017.drop_duplicates('Player')
stats_2017
```

```python
stats_2018 = df[df['Year']==2018]
stats_2018 = stats_2018.drop_duplicates('Player')
stats_2018
```

```python
stats_2019 = df[df['Year']==2019]
stats_2019 = stats_2019.drop_duplicates('Player')
stats_2019[stats_2019['Player']=='Carmelo Anthony']
```

```python
salary_2016 = pd.read_csv('data/2016_2017_salary.csv')
salary_2016 = salary_2016.sort_values('Player Name').drop(columns=['#', 'Unnamed: 3'])
salary_2016 = salary_2016.rename(columns={'Player Name': 'Player'})
```

```python
all_2016 = pd.merge(stats_2016, salary_2016, on='Player', how='left')
all_2016.to_csv('data/all_2016.csv')
```

```python
salary_2017 = pd.read_csv('data/2017_2018_salary.csv')
salary_2017 = salary_2017.sort_values('player').drop(columns=['#', 'Unnamed: 3'])
salary_2017 = salary_2017.rename(columns={'player': 'Player'})
```

```python
all_2017 = pd.merge(stats_2017, salary_2017, on='Player', how='left')
all_2017.to_csv('data/all_2017.csv')
```

```python
salary_2018 = pd.read_csv('data/2018_2019_salary.csv')
salary_2018 = salary_2018.sort_values('player').drop(columns=['#', 'Unnamed: 3'])
salary_2018 = salary_2018.rename(columns={'player': 'Player'})
```

```python
all_2018 = pd.merge(stats_2018, salary_2018, on='Player', how='left')
all_2018.to_csv('data/all_2018.csv')
```
