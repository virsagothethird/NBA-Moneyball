```python
drop_columns = ['Unnamed: 0', 'Unnamed: 0.1', 'Year', 'PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%',
       'STL%', 'BLK%', 'TOV%', 'USG%', 'OWS', 'DWS', 'WS/48', 'OBPM',
       'DBPM', 'BPM', 'VORP', 'FG%', '3P%', '2P%', 'eFG%', 'FT%', ' Player Salary in $ ']
drop_columns2 = ['Unnamed: 0', 'Unnamed: 0.1', 'Year', 'PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%',
       'STL%', 'BLK%', 'TOV%', 'USG%', 'OWS', 'DWS', 'WS/48', 'OBPM',
       'DBPM', 'BPM', 'VORP', 'FG%', '3P%', '2P%', 'eFG%', 'FT%', 'salary']
```

```python
all_2016 = all_2016.drop(columns=drop_columns)
all_2017 = all_2017.drop(columns=drop_columns2)
all_2018 = all_2018.drop(columns=drop_columns2)
```

```python
merged = all_2016.set_index('Player').join(all_2017.set_index('Player'),lsuffix='_2016', rsuffix='_2017',how='outer').reset_index()
```

```python
merged2 = merged.set_index('Player').join(all_2018.set_index('Player'), rsuffix='_2018',how='right').reset_index()
```

```python
merged2.info()
```

```python
merged2 = merged2.dropna(subset=['DRB_2017'])
merged2 = merged2.fillna(0)
```

```python
merged2.info()
```

```python
merged2.columns
```

```python
drop_columns = ['Player', 'Pos_2016', 'Tm_2016', 
       'WS_2016', 'ORB_2016', 'DRB_2016', 
       'Pos_2017', 'Tm_2017', 'WS_2017',
       'ORB_2017', 'DRB_2017', 'Pos', 'Age',
       'Tm', 'G', 'MP', 'WS', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', 'FT',
       'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
merged2_X = merged2.drop(columns=drop_columns)
merged2_y = merged2['WS']
```

```python
merged2_X
```

```python
X_train, X_test, y_train, y_test = tts(merged2_X,merged2_y)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf.score(X_test,y_test)
```

```python
predict = rf.predict(merged2_X)
```

```python
residuals = merged2_y - predict
```
